````python
# correlation between dct and pca
# fig:correlation_dct_pca
#
from scipy.fftpack import dct, idct

# DCT
X_bgr = cv2.imread('../data/kodim01.png')
X_bgr = BGR2YUV(X_bgr).reshape(1, 512, 768, -1)
X = Shrink(X_bgr[:,:,:,:1], win=8).reshape(1, 64, 96, 8, 8)
tX = dct(dct(X, axis=3, norm='ortho'), axis=4, norm='ortho')
tX = ZigZag().transform(tX.reshape(1, 64, 96,-1))
tX = tX.reshape(-1, 64)[:, :16]
tX = np.moveaxis(tX, 1, 0)
coef = np.corrcoef(tX)
plt.figure(figsize=(10,10))
plt.imshow(coef)
plt.colorbar()
plt.show()

# PCA
X_bgr = cv2.imread('../data/kodim01.png')
X_bgr = BGR2YUV(X_bgr).reshape(1, 512, 768, -1)
X = Shrink(X_bgr[:,:,:,:1], win=8)
pca = myPCA()
pca.fit(X)
tX = pca.transform(X)
tX = tX.reshape(-1, 64)[:, :16]
tX = np.moveaxis(tX, 1, 0)
coef = np.corrcoef(tX)
plt.figure(figsize=(10,10))
plt.imshow(coef)
plt.colorbar()
plt.show()
````

````python
# fit performance curve
#
from sklearn.ensemble import RandomForestRegressor

def performance_curve_fit(bpp_list, psnr_list):
    bpp = np.arange(50, 700, 10) / 200
    psnr = []
    for i in range(24):
        X, Y = np.array(bpp_list[i]), np.array(psnr_list[i])
        clf = RandomForestRegressor(n_estimators=300).fit(X.reshape(-1, 1), Y.reshape(-1, 1))
        psnr.append(clf.predict(bpp.reshape(-1, 1)))
    psnr = np.array(psnr)
    return bpp, np.mean(psnr, axis=0)
````

````python
# performance curve
def file_size(root):
    size = []
    q = [200, 160, 140, 120, 100, 90, 80, 70, 60, 50, 40.0,32.0,26.6,22.8,20.0,17.7,16.0,14.4,12.8,11.2,9.6,8.0,6.4,4.8,3.2]
    for i in range(24):
        tmp = []
        for j in range(len(q)):
            tmp.append(os.stat(root + str(i) + '_'  + str(j) + '.jpg').st_size)
        size.append(tmp)
    return np.array(size)

def plot(name):
    size = file_size('../result/'+name+'/')
    bpp = (size+64*4+18) * 8 / 512 / 768 
    with open('../result/psnr_'+name+'.pkl', 'rb') as f:
        psnr = pickle.load(f)
    with open('../result/our_color_20210128_'+name+'.pkl', 'wb') as f:
        pickle.dump({'psnr':psnr, 'bpp':bpp}, f)
    # jpeg color curve
    with open('../result/jpeg_color.pkl', 'rb') as f:
        jpg = pickle.load(f)
    jpg_bpp, jpg_psnr = performance_curve_fit( jpg['bpp'], jpg['psnr'])

    # our method color curve
    with open('../result/our_color_20210128_'+name+'.pkl', 'rb') as f:
        our = pickle.load(f)
    our_bpp, our_psnr = performance_curve_fit( our['bpp'], our['psnr'])
    
    BD_Performance(jpg_bpp, jpg_psnr, our_bpp, our_psnr)
    return jpg_bpp, jpg_psnr, our_bpp, our_psnr
    
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10))

print('YUV')
jpg_bpp, jpg_psnr, our_bpp, our_psnr = plot(name='our_yuv')
plt.plot(jpg_bpp, jpg_psnr, c='y',label='JPEG')
#plt.plot(our_bpp, our_psnr, label='Ours YUV')
print('PQR')
jpg_bpp, jpg_psnr, our_bpp, our_psnr = plot(name='our_pqr')
plt.plot(our_bpp, our_psnr, c='b',label='Ours PQR Color')
plt.xlabel('Bit Rate (bpp)')
plt.ylabel('PSNR (dB)')
plt.legend()
plt.show()

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10))
print('YUV')
jpg_bpp, jpg_psnr, our_bpp, our_psnr = plot(name='our_yuv1.4')
plt.plot(jpg_bpp, jpg_psnr, c='y',label='JPEG')
#plt.plot(our_bpp, our_psnr, label='Ours YUV')
print('PQR')
jpg_bpp, jpg_psnr, our_bpp, our_psnr = plot(name='our_pqr1.4')
plt.plot(our_bpp, our_psnr, c='b',label='Ours PQR Color')
plt.xlabel('Bit Rate (bpp)')
plt.ylabel('PSNR (dB)')
plt.legend()
plt.show()
````
